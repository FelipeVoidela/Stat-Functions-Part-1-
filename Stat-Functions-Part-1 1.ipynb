## Média (mean)

A média (ou valor médio) de uma lista de números é calculada somando todos os valores e dividindo pela quantidade de valores na lista.
Fórmula:


A média é calculada como $\text{Média} = \frac{1}{n} \sum_{i=1}^{n} x_i$.

```python
def media(vetor):
    return np.mean(vetor)
print("Média:", media(preco))
```

## Mediana (median)

A mediana é o valor do meio em uma lista de números ordenada. Se a lista tiver um número par de elementos, a mediana é a média dos dois valores do meio.


$$ \text{Mediana} = \begin{cases}
\text{valor do meio} & \text{se } n \text{ é ímpar} \\
\frac{x_{\frac{n}{2}} + x_{\frac{n}{2} + 1}}{2} & \text{se } n \text{ é par}
\end{cases} $$

```python
def mediana(vetor):
    return np.median(vetor)
print("Mediana:", mediana(preco))
```

## Max

O máximo e mínimo são simplesmente os maiores e menores valores na lista, respectivamente.

$\text{Máximo} = \max(x_1, x_2, \dots, x_n)$

$\text{Mínimo} = \min(x_1, x_2, \dots, x_n)$


```python
def maximo(vetor):
    return max(vetor)
print("Máximo:", maximo(preco))
```

## Min

```python
def minimo(vetor):
    return min(vetor)
print("Mínimo:", minimo(preco))
```

## Moda (mode)

A moda é o valor que aparece com mais frequência na lista.

```python
def moda(vetor):
    valores, contagens = np.unique(vetor, return_counts=True)
    indice_moda = np.argmax(contagens)
    return valores[indice_moda]

print("Moda:", moda(preco))
```

## Desvio Padrão (standard deviation - stdev)

O desvio padrão mede a dispersão dos valores em relação à média. Quanto maior o desvio padrão, mais dispersos estão os valores.

$\text{Desvio Padrão} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - \text{Média})^2}$


```python
def desvio_padrao(vetor):
    return np.std(vetor)
print("Desvio Padrão:", desvio_padrao(preco))
```

## Variância (variance)

A variância é o quadrado do desvio padrão e é uma medida da dispersão dos valores.

$\text{Variância} = \frac{1}{n} \sum_{i=1}^{n} (x_i - \text{Média})^2$


```python
def variancia(vetor):
    return np.var(vetor)
print("Variância:", variancia(preco))
```
